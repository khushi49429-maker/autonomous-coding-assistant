```python
import requests
import time

BASE_URL="http://127.0.0.1:8000"

start=time.time()
response=requests.get(f"{BASE_URL}/health")
end=time.time()

print("Status Code:",response.status_code)
print("Response:",response.text)
print("Response Time:",round(end-start,3),"seconds")
```

## Test Case 1

**API Tested:** GET `/health`

**Expected Result:** Backend should return a successful response.

**Actual Result:** Backend responded successfully.

**Status:** PASS


```python
payload={"prompt":"Write a Python program to find the factorial of a number."}
response=requests.post(f"{BASE_URL}/generate-code",json=payload)
print("Status Code:",response.status_code)
print(response.text)
```
