### Project Description

CodeMentor AI is an intelligent coding assistant developed using Google Gemini AI, FastAPI, MySQL, and GitHub REST API.

The application assists developers by generating code, explaining source code, reviewing code quality, detecting bugs, suggesting fixes, and analyzing GitHub repositories through OAuth authentication.

- Evaluate GitHub Repository Access
- Evaluate GitHub File Analysis

````python
from google import genai
from dotenv import load_dotenv
import os

load_dotenv()

client = genai.Client(
    api_key=os.getenv("GEMINI_API_KEY")
)
````

````python
def ask_ai(prompt):
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return response.text
````

# Test Case 1
## AI Code Generation

````python
prompt = """
Write a Python program to calculate factorial using recursion.
"""

print(ask_ai(prompt))
````

# Test Case 2
## Code Explanation

````python
prompt = """
Explain the following Python code.

for i in range(5):
    print(i)
"""

print(ask_ai(prompt))
````
